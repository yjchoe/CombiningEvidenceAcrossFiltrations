# Reproducing E-processes for Testing Exchangeability

Testing whether a data sample is exchangeable (e.g., IID) is a nontrivial problem for which different e-processes are constructed using different filtrations.

This notebook includes test runs with various types of simulated data (null & alternative).

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import seaborn as sns
import os

import ecombine as ec
```

```python
# basic figure styles
ec.set_theme()

# rng for data generation and inference methods, separately
rng_data = np.random.default_rng(1234)
rng = np.random.default_rng(5678)

# plots saved here
plots_dir = "./plots/exch/reprod"
os.makedirs(plots_dir, exist_ok=True)
mpl.rc("savefig", format="png", dpi=300)
```

Unless specified otherwise, we fix the size of the data to be $n=10,000$.

```python
n = 10000
```

## i.i.d. data (null)

First, we generate data from a null distribution:
$$
X_1, \dotsc, X_t \overset{i.i.d.}{\sim} \mathrm{Bernoulli}(p).
$$

```python
p = 0.1

x_iid = ec.data.exch.generate_binary_iid(p=p, size=n, rng=rng_data)

ax = sns.histplot(x_iid)
ax.set_title(f"Data: i.i.d. Bernoulli({p})")
None
```

We first compute and plot various e-processes for exchangeability:

```python
ec.plot_eprocesses_exch(
    x_iid, 
    title=f"E-processes for testing exchangeability: i.i.d. Ber({p})",
    rng=rng,
)
None
```

## Non-iid exchangeable data

We construct exchangeable binary sequences $X_1, X_2, \dotsc$ using i.i.d. Bernoulli sequences $B_0, B_1, B_2, \dotsc \sim Ber(p)$ such that
$$
X_t = B_0 \wedge B_t, \quad \forall t \geq 1.
$$
This is an exchangeable but non-iid data.

```python
p = 0.5
x_exch = ec.data.exch.generate_binary_exch_noniid(
    p=p, 
    size=n, 
    rng=rng_data,
)

ax = sns.histplot(x_exch)
ax.set_title(f"Data: Exch. and non-i.i.d.")
None
```

```python
ec.plot_eprocesses_exch(
    x_exch, 
    title=f"E-processes for testing exchangeability: Exch. & non-i.i.d. binary data",
    rng=rng,
)
None
```

## Changepoint data (alternative)

Now we consider data sampled from an alternative distribution. The first is a change point data, for which the conformal test martingales are known to perform well:

```python
p, q = 0.1, 0.5
# p, q = 0.01, 0.05

x_change = ec.data.exch.generate_binary_changepoint(
    p=p, q=q, size=n, rng=rng_data)

ax = sns.histplot(x_change)
ax.set_title(f"Data: Change from Ber({p}) to Ber({q})")
None
```

```python
ec.plot_eprocesses_exch(
    x_change, 
    title=f"E-processes for testing exchangeability: Ber({p}) then Ber({q})",
    rng=rng,
)
None
```

## First-order Markov data (alternative 2)

Next, we look at first-order Markov data, for which the UI e-process is consistent/powerful:

```python
# p_kj: transition prob from j to k
# p_10, p_11 = 0.5, 0.1
p_10, p_11 = 0.05, 0.01

x_markov = ec.data.exch.generate_binary_markov(
    p_10=p_10, p_11=p_11, size=n, rng=rng_data)

ax = sns.histplot(x_markov)
ax.set_title(f"Data: Markov({p_10}, {p_11})")
None
```

```python
ec.plot_eprocesses_exch(
    x_markov, 
    title=f"E-processes for testing exchangeability: Markov({p_10}, {p_11})",
    rng=rng,
)
None
```

Finally, we show include some higher-order Markov data for which neither e-process is known to be optimal.

```python
# 2nd order
x_0011 = np.tile([0, 0, 1, 1], n // 4)
x_0011[:16]
```

```python
ec.plot_eprocesses_exch(
    x_0011, 
    title=f"E-processes for testing exchangeability: Markov-2 (0011)",
    rng=rng,
)
None
```
